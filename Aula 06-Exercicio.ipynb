```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.cross_validation import cross_val_score
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('acessos_buscas2.csv')
df
```

```python
x = df[['home', 'busca', 'logado']]
y = df['comprou']
x = pd.get_dummies(x)
```

### Multinomial

```python
modelo = MultinomialNB()
k = 10
scores = cross_val_score(modelo, x, y, cv=k)
print('Scores: ', scores)
taxa_de_acerto = np.mean(scores)
print('Taxa de Acerto: ', taxa_de_acerto)
desvio = scores.std()
print('Desvio: ', desvio)
```

### AdaBoost

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import cross_val_score
from scipy.stats import bayes_mvs
```

```python
modelo_ada = AdaBoostClassifier(n_estimators=100)
k = 10
scores = cross_val_score(modelo_ada, x, y, cv=k)
print('Taxa de acerto:', scores.mean())
print('Desvio padrão', scores.std())
print()
print('Int. Confiança: ', bayes_mvs(scores, 0.95))
```
